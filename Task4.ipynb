```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')


# Year and runs data
years = [2021, 2022, 2023, 2024, 2025]
virat_runs = [295, 400, 310, 280, 270]
ruturaj_runs = [220, 110, 190, 210, 180]
ishan_runs = [115, 470, 205, 140, 250]
abhishek_runs = [90, 60, 100, 255, 278]
yashasvi_runs = [45, 78, 130, 190, 188]


plt.plot(years, virat_runs, color='blue', linestyle='--', marker='o', linewidth=2, markersize=8, label='Virat Kohli')
plt.plot(years, ruturaj_runs, color='orange', linestyle='-.', marker='s', linewidth=2, markersize=8, label='Ruturaj Gaikwad')
plt.plot(years, ishan_runs, color='green', linestyle=':', marker='^', linewidth=2, markersize=8, label='Ishan Kishan')
plt.plot(years, abhishek_runs, color='red', linestyle='-', marker='d', linewidth=2, markersize=8, label='Abhishek Sharma')
plt.plot(years, yashasvi_runs, color='purple', linestyle='--', marker='*', linewidth=2, markersize=10, label='Yashasvi Jaiswal')


plt.title("T20 Runs by Year for Indian Cricketers")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
