```python
# ----------- 1. Top 10 String Functions -----------
print("----- Top 10 String Functions -----")
s = " Hello Python World "

print("Original String:", s)

# 1. strip()
print("1. strip():", s.strip())

# 2. lower()
print("2. lower():", s.lower())

# 3. upper()
print("3. upper():", s.upper())

# 4. replace()
print("4. replace():", s.replace("Python", "Programming"))

# 5. split()
print("5. split():", s.split())

# 6. join()
words = ['Join', 'these', 'words']
print("6. join():", ' '.join(words))

# 7. find()
print("7. find():", s.find("Python"))

# 8. count()
print("8. count():", s.count("o"))

# 9. startswith()
print("9. startswith(' Hello'):", s.startswith(" Hello"))

# 10. endswith()
print("10. endswith('World '):", s.endswith("World "))


# ----------- 2. Bitwise Operators -----------
print("\n----- Bitwise Operators -----")
a = int(input("Enter first number (a): "))
b = int(input("Enter second number (b): "))

print("a =", a, ", b =", b)
print("a & b =", a & b)      # AND
print("a | b =", a | b)      # OR
print("a ^ b =", a ^ b)      # XOR
print("~a =", ~a)            # NOT
print("a << 1 =", a << 1)    # Left Shift
print("a >> 1 =", a >> 1)    # Right Shift


# ----------- 3. Assignment Operators (with input) -----------
print("\n----- Assignment Operators (with input) -----")
x = int(input("Enter a number for x: "))

print("Initial x =", x)

y = int(input("Enter number to add to x: "))
x += y
print("x += y -->", x)

z = int(input("Enter number to subtract from x: "))
x -= z
print("x -= z -->", x)

p = int(input("Enter number to multiply with x: "))
x *= p
print("x *= p -->", x)

q = int(input("Enter number to divide x by: "))
x /= q
print("x /= q -->", x)

x %= 2
print("x %= 2 -->", x)

x = int(input("Re-enter x for exponentiation: "))
x **= 2
print("x **= 2 -->", x)

x //= 3
print("x //= 3 -->", x)


# ----------- 4. if - elif - else -----------
print("\n----- if - elif - else Example -----")
num = int(input("Enter any number: "))

if num > 0:
    print("The number is Positive.")
elif num == 0:
    print("The number is Zero.")
else:
    print("The number is Negative.")


# ----------- 5. File Handling Operations -----------
print("\n----- File Handling Operations -----")

filename = "sample.txt"  # File name

# 1. Write to the file
with open(filename, "w") as f:
    f.write("This is the first line written to the file.\n")
    f.write("Python file handling is easy!\n")
print("✔ File written successfully.")

# 2. Read the file
print("\nReading the file after write:")
with open(filename, "r") as f:
    content = f.read()
    print(content)

# 3. Append to the file
with open(filename, "a") as f:
    f.write("This line is appended to the file.\n")
print("✔ File appended successfully.")

# 4. Read again after appending
print("\nReading the file after append:")
with open(filename, "r") as f:
    content = f.read()
    print(content)
```
