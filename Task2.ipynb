```python
print("----- List Methods -----")
fruits = ["apple", "banana", "cherry", "banana"]

print("Original List:", fruits)

# append()
fruits.append("orange")
print("append():", fruits)

# insert()
fruits.insert(1, "mango")
print("insert():", fruits)

# remove()
fruits.remove("banana")
print("remove():", fruits)

# pop()
fruits.pop()
print("pop():", fruits)

# index()
print("index():", fruits.index("mango"))

# count()
print("count('banana'):", fruits.count("banana"))

# sort()
fruits.sort()
print("sort():", fruits)

# reverse()
fruits.reverse()
print("reverse():", fruits)

# copy()
fruits_copy = fruits.copy()
print("copy():", fruits_copy)

# clear()
fruits.clear()
print("clear():", fruits)



print("\n----- Tuple Methods -----")
numbers = (10, 20, 30, 20, 40, 20)

print("Tuple:", numbers)

# count()
print("count(20):", numbers.count(20))

# index()
print("index of 30:", numbers.index(30))

# Note: Tuples are immutable, so no add/remove methods.



print("\n----- Set Methods -----")
colors = {"red", "green", "blue"}

print("Original Set:", colors)

# add()
colors.add("yellow")
print("add():", colors)

# update()
colors.update(["orange", "pink"])
print("update():", colors)

# remove()
colors.remove("red")
print("remove():", colors)

# discard()
colors.discard("purple")
print("discard():", colors)

# pop()
removed_color = colors.pop()
print("pop():", removed_color)
print("After pop():", colors)

# copy()
colors_copy = colors.copy()
print("copy():", colors_copy)

# clear()
colors.clear()
print("clear():", colors)
```
